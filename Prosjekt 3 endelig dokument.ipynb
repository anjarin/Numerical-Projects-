<b> Innledning

I dette prosjektet skal vi finne damptrykkurven til vann ved hjelp av van der Waals ligning og Clapeyrons ligning. Damptrykkurven beskriver trykket ved likevekten til gass- og væske-fasen til stoff, og går fra trippelpunktet til kritisk punkt til stoffet. I første oppgave ser vi hovedsakelig på van der Waals ligning og løser denne med Newtons metode. I andre oppgave vil vi fokusere på Clapeyrons ligning som kan løses med numerisk integrasjon, i vårt tilfelle med Simpsons metode. I begge oppgavene vil vi sammenligne med eksperimentelle verdier og se hvor godt ligningene og de numeriske metodene klarer å beskrive damptrykkurven.

<b> Teori/motivasjon

For å finne damptrykkurven til $H_2O$ kan vi bruke at stigningstallet til damptrykket er gitt ved

$\frac{dp}{dT}=\frac{\Delta S}{\Delta V} \tag{1}$

der $\Delta S = S_g-S_v$ er entropien og $\Delta V = V_g-V_v$ er volumet når en viss stoffmengde omdannes fra væske til gass, ved trykk $p$ og temperatur $T$ . Dette er Clapeyrons ligning som kan integreres for å finne p(T) slik:

$p(T) = p_t +\int_{T_t}^{T}\frac{L(T)dT}{T\Delta V(T)} \tag{2}$

Her har vi brukt at $\Delta S = \frac{L(T)}{T}$, der $L$ er fordampningsvarmen. Siden vi vet at damptrykkurven til vann går fra trippelpunktet til kritisk punkt, bruker vi trippelpunketet $T_t$ som referanse.
For å finne p(T) må man først finne ligninger som beskriver $L(T)$, samt $\Delta V(T)$ for så å kunne integrere ligningen numerisk. For å finne en analytisk løsning for damptrykkurven må vi gjøre et par antakelser. For det første kan vi anta at fordampningsvarmen $L$ er konstant i et lite temperaturområde rundt et valgt punkt $(p_0,T_0)$. Hvis vi er langt nok unna kritisk punkt vil gassvolumet være mye større enn væskevolumet. Sammen med å anta ideell gass får vi da $\Delta V\approx V_g = \frac{nRT}{p} $. Hvis vi nå integrerer clapeyrons ligning får  vi

$p(T) = p_0exp[\frac{L}{R}(\frac{1}{T_0}-\frac{1}{T})] \tag{3}$

En tilnærming til damptrykkurven kan finnes med van der Waals ligning, som er en forbedret versjon av den ideelle gassloven.Ideell gasslov er gitt ved:

$p = \frac{nRT}{V} \tag{4}$

der $n$ er antall mol, R er universell gasskonstant, $V$ er volumet og $T$ er temperatur.
Van der Waals ligning er gitt ved

$p = \frac{RT}{V-b}-\frac{a}{V^2} \tag{5}$

dersom det er 1 mol med stoff. Siden 1 mol med $H_2O$ vil studeres gjennom hele prosjektet er det kun denne ligningen som brukes videre. Parametrene $a$ og $b$ justerer respektivt for de svake tiltrekkende kreftene mellom molekyler som gir et redusert trykk og at molekyler ikke er punktpartikler, og dermed har tilgjengelig volum som er mindre enn det totale volumet $V$. $a$ og $b$ er empiriske og fastlegges vanligvis med to av de tre størrelsene $T_c$, $p_c$ og $V_c$ i kritisk punkt.Van der Waals ligning er kun er en forbedring i forhold til ideell gasslov med tanke på å finne damptrykkkurven, og ikke eksakt.

En av de viktige feilene til Van der Waals ligning er at for $T<T_c$ vil det gi ufysiske verdier for volumer mellom $V_v$ og $V_g$. Det burde gi en konstant koeksistenskurve, men det vil i stedet bli en kurve som krysser denne linjen i tre punkt. For å finne trykket til koeksistenskurven, $p_0$, kan Maxwells konstruksjon benyttes. Av de tre punktene  er den med lavest volum $V_v$, den med høyest volum er $V_g$ og punktet i midten er benevnet $V_0$. $V_0$ er plassert slik slik at arealet mellom van der Waals kurven og likevektskurven mellom $V_v$ og $V_0$ er det samme som arealet mellom likevektskurven og van der Waals kurven mellom $V_0$ og $V_v$. Dette gir dermed to ligninger for $V_v$ og $V_g$ som i dette prosjektet skal løses med Newtons metode. Ligningene er gitt ved

$$\frac{RT}{V_g-b}-\frac{a}{V_g^2}-\frac{RT}{V_v-b} + \frac{a}{V_v^2}  \tag{6}$$

og 

$$\frac{RT}{V_g-V_v}log(\frac{V_g-b}{V_v-b})-\frac{a}{V_gV_v} -\frac{RT}{V_g-b}+\frac{a}{V_g^2} \tag{7}$$

der alle parametrene er nevnt i teorien over (Støvneng, et al. 2023. s.5).



## <center>  Oppgave 1: van der Waals tilstandsligning og damptrykk

<b> 1a

Van der Waals ligning er gitt ved ligning (4). Parametrene $a$ og $b$ er empiriske og kan bestemmes med denne relasjonen for kritisk temperatur og trykk:

$$(\frac{\partial p}{\partial V})_c = (\frac{\partial^2 p}{\partial V^2})_c = 0$$

Ut ifra disse realsjonene finner vi

$$V_c = 3b, T_c = \frac{8a}{27Rb}, p_c = \frac{a}{27b^2}$$

der $V_c$, $T_c$ og $p_c$ er henholdsvis de kritisk verdiene for volum, trykk og temperatur. $R$ er den universelle gasskonstanten.

Hvis vi så bruker ligningene for $p_c$ og $T_c$ kan vi finne utrykk for $a$ og $b$:

$$a = \frac{27R^2T_c^2}{64p_c}, b = \frac{RT_c}{8p_c}$$

Det kritiske molare volumet med disse ligningene blir $V_c = \frac{3RT_c}{8p_c}$, som med verdier for $T_c$ og $p_c$ gir $V_c = 9.144 * 10^{-5} m^3 = 91.44 ml$. Eksperimentell verdi for $V_c = 55.948 ml$. Det er klart at Van der Waals tilstandsligning ikke gir samsvar med eksperimentell verdi, og da ikke kan brukes for kvantitative resultater (Støvneng, et al. 2023. s. 4).



<b> 1b

Trykket kan plottes som funksjon av volumet med konstant temperatur ved hjelp av van der Waals tilstandsligning. Under brukes temperaturen $T = T_c$. 

```python
# Importerer nødvendige biblioteker
import numpy as np
import pandas as pd
import scipy.optimize
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [10, 6]
```

```python
# Oppretter nødvendige konstanter
R = 8.314       
Tc = 647.096
pc = 22.064*10**6
a = 27*R**2*Tc**2/(64*pc)
b = R*Tc/(8*pc)
```

```python
def p(V, T):
    '''
    Tar inn et volum og regner ut trykket til 1 mol H2O ved kritisk temperatur
    ---------------------
    Input:
        V: volum til fluidet (m^3)
        T: temperatur (K) 
    Return:
        P: trykk til fluidet (bar)
    '''

    #regner ut trykk med Van der Waals ligning
    P = (8.314*T/(V-b) - a/V**2)/(10**5)
    return P


#plotter trykk vs volum ved 647 K
v = np.linspace(75, 300, 225)*10**-6
plt.title('1 mol $H_2O$, van der Waals isoterm 647 K')
plt.xlabel('Volum (ml)')
plt.ylabel('Trykk (bar)')
plt.plot(v*10**6, p(v, 647.096))
plt.show()
```

Dette gir mening fordi ved kritisk temperatur vil trykkurven ikke passere gjennom damptrykkurven, men heller tangere den ved kritisk punkt. Det betyr og at det ikke er noen ufysiske punkter i denne grafen. 

Dersom Van der Waals ligning brukes til å finne en isoterm i et pV diagram, så vil den gi ufysiske vedrier for $T$ under $T_c$. For disse temperaturene vil trykket mellom $V_v$ og $V_g$  være konstant, siden dette beskriver faseovergangen mellom vsæke og gass (Støvneng, et al. 2023. s.4).

<b> 1c

Newtons metode skal nå implementeres for en variabel og brukes til å løse en testligning. Newtons metode for en variabel løser en ligning på formen $f(x) = 0$ for x. Den kan formuleres slik:

$$x_{n+1} = x_n + \frac{f(x_n)}{f'(x_n)}$$

der $x_n$ er startverdi for en rekursjon. Rekken {$x_n$} vil konvergere mot roten til $f(x)$ dersom startverdien $x_0$ er nerme nok til roten.



```python
def Newton1(x_0, func, funcD, tollDiff):
    '''
    Løser ligning numerisk med Newtons metode, og returnerer løsning 
    ---------------------
    Input:
        x_0: Startverdien for iterasjon (K)
        func: Funksjonen som brukes for iterasjon
        funcD: Derivert av func
        tollDiff: toleransen for hvor liten forskjell en iterasjon skal gi før løkken brytes
    Return:
        x: array med itererte x verdier, siste verdi er numerisk løsning (K)
    '''
    
    # Oppretter liste med startverdi for x
    x = []
    x.append(x_0)
    # Setter maks antall iterasjoner
    for i in range(20):
        
        # Sjekker for å kjøre minst 3 ganger
        if i > 2:
            
            # Sjekker for stor nok endring mellom forrige 2 iterasjoner til å iterere igjen
            if (abs(x[-1] - x[-2])) > tollDiff:
                
                # Regner ut neste verdi
                x.append(x[-1]-func(x[-1])/funcD(x[-1]))
            else:
                break
        else:
            # Regner ut neste verdi
            x.append(x[-1]-func(x[-1])/funcD(x[-1]))
            
    return np.array(x)


def func(Tc):
    '''
    Testfunksjon
    ---------------------
    Input:
        Tc: kritist temperatur
    Return:
        verdien av funksjonen i Tc
    '''
    try:
        return np.sinh(2/Tc)**2-1
    except:
        return 0

def funcD(Tc):
    '''
    Derivert av testfunksjonen
    ---------------------
    Input:
        Tc: kritist temperatur
    Return:
        verdien av den deriverte til funksjonen i Tc
    '''
    try:
        return 2*np.sinh(2/Tc)*np.cosh(2/Tc)*(-2/Tc**2)
    except:
        return 1



print('Nå skal det undersøkes hvordan startverdi påvirker Newtons metode.')
print('Dette er svaret metoden konvergerer mot for verdier mellom -5 og 5, (ikke 0 fordi det gir division by zero error) \n')

newtonTest = Newton1(1, func, funcD, 0.0000000001)

newtonTestList = []
for i in range(-5,6):
    if i != 0:
        newtonTestList.append(Newton1(i, func, funcD, 0.0001)[-1])
print(newtonTestList)
```

Newtons metode er veldig praktisk for å løse en ligning, og konvergerer raskt mot et bra svar, gitt en god start verdi. Ikke alle startverdier fører til at metoden konvergerer mot riktig svar.  Ut ifra svarene for startverdi mellom -5 og 5 ses det at man får riktig svar for startverdier [-4, 1, 2, 3]. Startverdien påvirker numerikken i relativt liten grad hvis startverdien er god nok til å konvergere, fordi Newtons metode konvergerer raskt. Om startverdien ikke er nærme nok derimot, er det ikke sikkert at metoden konvergerer, og dette kan føre til overflow, og division by zero feil. 

<b> 1d

Fordi den analytiske løsningen til testfunksjonen er kjent, kan feilen beregnes, og dermed er det mulig å finne konvergensgraden numerisk. Dette gjøres ved hjelp av følgen $p_i$, som konvergerer mot convergensgraden q. Rekken $p_i$ er definert som 

$$p_i = \frac{log(e_i/e_{i-1})}{log(e_{i-1}/e_{i-2})} $$

der $e_k$ er feilen for k-te steg
(Støvneng, et al. 2023. s. 8).

```python
def error(analVerdi, numLos):
    '''
    Tar inn en analytisk verdi, og en array, og regner ut feilen i de numeriske verdiene
    ---------------------
    Input:
        analVerdi: analytisk løsning 
        numLos: numeriske løsninger
    Return:
        e: array med feiler 
    '''
    # Oppretter tom array
    e = np.zeros(len(numLos))
    
    # Itererer gjennom numeriske verdier, og legger absolutt feil til arrayet
    for i in range(len(numLos)):
        e[i] = abs(analVerdi-numLos[i])
    return e


def grad(e):
    '''
    Regner ut forholdet mellom logaritmen til endringen in feilen i to følgende steg for å bestemme konvergensgraden til Newtons metode
    ---------------------
    Input:
        e: array med feiler
    Return:
        p: forholdet mellom logaritmen til endringen i feil mellom 2 steg og de forrige 2 stegene
    '''
    # Oppretter tom array
    p = np.zeros(len(e)-2)
    
    # Itererer og regner ut p verdier
    for i in range(2,len(e)):
        p[i-2] = (np.log(e[i]/e[i-1]))/(np.log(e[i-1]/e[i-2]))
    return p


# Regner ut analytisk verdi, og plotter beregnet feil
TcA = 2/(np.log(1+np.sqrt(2)))
print('Analytisk verdi: ', TcA)
plt.title('Feil med Newtons metode')
plt.plot(error(TcA, newtonTest))
plt.yscale("log")
plt.ylabel('Feil (K)')
plt.xlabel('i')
plt.show()


# Plotter p-rekken
plt.plot(grad(error(TcA, newtonTest)))
plt.title('$p_i$ verdier')
plt.ylabel('$p_i$')
plt.xlabel('i')
plt.show()
```

Fordi grafen til feilen minker raskere enn en lineær funksjon, kan er det tydelig at Newtons metode konvergerer superlineært. For å bestemme nøyaktig konvergensgrad må p-rekken studeres.

Det er kjent at p-rekken vil konvergere mot q for en gitt metode. Ut ifra grafen ser man at følgen $p_i$ konvergerer mot 2, som viser at Newtons metode har konvergensgrad lik 2.

<b> 1e

For å finne damptrykkurven numerisk må ligningsettet gitt ligningene (6) og (7) løses for $V_v$ og $V_g$. I denne oppgaven blir Newtons metode med to variabler implementert for å finne molart gassvolum, og molart væskevolum på damptrykkkurven.

```python
def Newton2(T, x_0, y_0, F, G, Fdx, Fdy, Gdx, Gdy, tollDiff):
    '''
    Løser et ligningsett med 2 ukjente med newtons metode
    ---------------------
    Input:
        T: Temperaturen ligningsettet løses for
        x_0: Startverdien til x
        y_0; Startverdien til y
        F: Den første funksjonen i systemet
        G: Den andre funksjonen i systemet
        Fdx: F derivert med hensyn på x
        Fdy: F derivert med hensyn på y
        Gdx: G derivert med hensyn på x
        Gdy: G derivert med hensyn på y
        tollDiff: toleransen for hvor liten forskjell en iterasjon skal gi før løkken brytes
    Return:
        X: Alle x verdier funnet gjennom iterasjon, den siste verdien er numerisk løsning
        Y: Alle y verdier funnet gjennom iterasjon, den siste verdien er numerisk løsning
        
    '''
    #oppretter startverdier og lister for x og y verdier
    X = [x_0]
    Y = [y_0]
    #itererer ikke mer en 20 ganger
    for i in range(20):
        #itererer minst 3 ganger
        if i > 2:
            #sjekker om forbedringen for hver iterasjon er over toleransen
            if X[-1]**2 + Y[-1]**2 > tollDiff**2:
                #lager array med verdier for Jacobi matrisen, og funksjonen
                J = np.array([[Fdx(X[i], Y[i], T), Fdy(X[i], Y[i], T)],[Gdx(X[i], Y[i], T), Gdy(X[i], Y[i], T)]])
                f = np.array([F(X[i], Y[i], T), G(X[i], Y[i], T)])
                
                #løser ligningsettet for delta x, og finner nye verdier for X og Y
                x = np.linalg.solve(J, -f)
                X.append(X[-1] + x[0])
                Y.append(Y[-1] + x[1])
            #Hvis endringen mellom 2 iterasjoner er mindre en tollDiff så stopper løkken
            else:
                break
        else:
            #regner ut minst 3 steg
            J = np.array([[Fdx(X[i], Y[i], T), Fdy(X[i], Y[i], T)],[Gdx(X[i], Y[i], T), Gdy(X[i], Y[i], T)]])
            f = np.array([F(X[i], Y[i], T), G(X[i], Y[i], T)])
    
            x = np.linalg.solve(J, -f)
            X.append(X[-1] + x[0])
            Y.append(Y[-1] + x[1])
    
    return X, Y
```

```python

# Vg = x, Vv = y
def F(Vg, Vv, T):
    '''
    En av funksjonene i ligningsettet
    ---------------------
    Input:
        Vg: Gassvolum
        Vv: Væskevolum
        T: Temperatur
    Return:
        Verdien til en av sidene til ligningsettet meg gitte verdier
    '''

    return R*T/(Vg-b)-a/Vg**2-R*T/(Vv-b)+a/Vv**2

def Fdx(Vg, Vv, T):
    '''
    F derivert med hensyn på Vg
    ---------------------
    Input:
        Vg: Gassvolum
        Vv: Væskevolum
        T: Temperatur
    Return:
        Verdien til F derivert med hensyn på Vg meg gitte verdier
    '''

    return -R*T/(Vg-b)**2+2*a/Vg**3

def Fdy(Vg, Vv, T):
    '''
    F derivert med hensyn på Vv
    ---------------------
    Input:
        Vg: Gassvolum
        Vv: Væskevolum
        T: Temperatur
    Return:
        Verdien til F derivert med hensyn på Vv meg gitte verdier
    '''
    
    return R*T/(Vv-b)**2-2*a/Vv**3

def G(Vg, Vv, T):
    '''
    En av funksjonene i ligningsettet
    ---------------------
    Input:
        Vg: Gassvolum
        Vv: Væskevolum
        T: Temperatur
    Return:
        Verdien til en av sidene til ligningsettet meg gitte verdier
    '''

    return R*T/(Vg-Vv)*np.log((Vg-b)/(Vv-b)) - a/(Vv*Vg)-R*T/(Vg-b)+a/Vg**2

def Gdx(Vg, Vv, T):
    '''
    G derivert med hensyn på Vg
    ---------------------
    Input:
        Vg: Gassvolum
        Vv: Væskevolum
        T: Temperatur
    Return:
        Verdien til G derivert med hensyn på Vg meg gitte verdier
    '''

    return -R*T*np.log((Vg-b)/(Vv-b))/(Vg-Vv)**2 + R*T/((Vg-b)*(Vg-Vv)) + a/(Vg**2*Vv) + R*T/(Vg-b)**2 - 2*a/Vg**3

def Gdy(Vg, Vv, T):
    '''
    G derivert med hensyn på Vv
    ---------------------
    Input:
        Vg: Gassvolum
        Vv: Væskevolum
        T: Temperatur
    Return:
        Verdien til G derivert med hensyn på Vv meg gitte verdier
    '''
   
    return R*T*np.log((Vg-b)/(Vv-b))/(Vg-Vv)**2 - R*T/((Vg-Vv)*(Vv-b)) + a/(Vg*Vv**2)
```

```python
# Opretter array for temperaturer og væske- og gass-volum
T = np.arange(274, 648, 1)
Vg = np.zeros(len(T))
Vv = np.zeros(len(T))
VgTrue = np.zeros(len(T))
VvTrue = np.zeros(len(T))

# Oppretter startverdier
Vg_0 =     12.6*10**-3
Vv_0 =     35.7*10**-6

# Itererer gjennom T verdier og regner ut Vv og Vg
for i in range(len(T)):
    
    # Hvis det ikke er første iterasjon endres startverdi for Newtons metode til forrige svar
    if i > 0:
        Vg_0 = Vg[i-1]
        Vv_0 = Vv[i-1]
    c, d = Newton2(T[i], Vg_0, Vv_0, F, G, Fdx, Fdy, Gdx, Gdy, 0.0001)
    Vg[i] = c[-1]
    Vv[i] = d[-1]

    
# Plotter Vv og Vv som funksjon av T
plt.plot(T, Vg*1000, label = "Gass")
plt.plot(T, Vv*1000, label = "Væske")
plt.legend()
plt.xlabel('temperatur (Kelvin)')
plt.ylabel('molart volum ($l/mol$)')
plt.show()
```

Ut ifra plottet for $V_g$ og $V_v$ , så ses det at når temperaturen går mot $T_c$ så vil forskjellen mellom $V_g$ og $V_v$ gå mot null. Dette gir mening fordi i kritisk punkt vil det ikke lenger være forskjell mellom væske og gassfase. Og dermed ikke noe forskjell mellom væskevolum og gassvolum.

Vi skal nå se på hva som skjer dersom man bruker van der Waals ligning til å finne P(V) for temperaturer under $T_c$. Vi valgte $T = 500 K$.

```python
# Henter ut verdier for Vv, Vg og L som funskjon av T
mydata_gass = pd.read_csv("Vitber_data_2a.csv", skiprows= 3, usecols=[1,2, 8], engine='python')                # ( Engineering ToolBox, (2003). Saturated Steam - Properties for Pressure in Bar).
mydata_vann = pd.read_csv("Vitber_data_2a.csv", skiprows= 2,skipfooter=31,  usecols=[12,13], engine='python')  # ( Engineering ToolBox, (2004). Water - Specific Volume vs. Temperature )


# Vanndamp, gassform
temp_gass = mydata_gass["(oC)"].values
Vg_eksp = mydata_gass["(m3/kg)"].values

# Vann, væskeform
temp_vann = mydata_vann["[°C]"].values
Vv_eksp = mydata_vann["[cm3/g]"].values

# Latent varme, L:
latent_varme = mydata_gass["(kJ/kg).2"].values


# Gjøre om til SI-enheter:
temp_gass += 273.15                # Gjør om fra Celsius til Kelvin
Vg_eksp = (Vg_eksp/1000)*18.016    # Gjør om til m^3/mol, bruker: m_H2O: 18.016g/mol

temp_vann += 273.15                # Gjør om fra Celsius til Kelvin
Vv_eksp = (Vv_eksp*10**-6)*18.016  # Gjør om til m^3/mol

latent_varme = latent_varme*18.016 # Gjør om til J/mol

print(f"temp_gass går fra T = {temp_gass[0].round()} til T={temp_gass[-1]}")
print(f"temp_vann går fra T = {temp_vann[0]} til T={temp_vann[-1]}")
```

<b> 2b

Siden vi kun har L(T), Vg(T) og Vv(T) for et visst antall verdier for T, er det vanskelig å integrere van der Waals ligning. Derfor vil vi finne tilnærmede funskjoner som beskriver L, Vg og Vv og som kan deriveres slik at disse kan brukes i Simpsons metode. 

```python
#---------------------------- Latent varme ----------------------------------------------

# Finner funskjon som gir god tilpasning til data for latent varme, L:
def func_l(x, a, b, c, d):                        # Denne skal optimaliseres
    return a+b*x+c*x**2+abs(647.096-x)**d         # Tc = 647.096K

l_opt, l_cov = scipy.optimize.curve_fit(func_l, temp_gass, latent_varme)



# Plot latent varme
plt.plot(temp_gass, latent_varme)                        # Eksperimentell måleserie
plt.plot(temp_gass, func_l(temp_gass, *l_opt), "r--")    # Curve fit
plt.title("Latent varme (L) H2O")
plt.legend(["Eksperimentell måleserie", "L(t) = a+bt+ct^2+|Tc-t|^d"])
plt.xlabel("Temperartur (K)")
plt.ylabel("L ( J/mol )")
plt.text(430, 42000, f"a={l_opt[0].round(2)},b={l_opt[1].round(2)},c={l_opt[2].round(2)},d={l_opt[3].round(2)}")
plt.show()


#------------------------------ Vv ---------------------------------------------------

# Finner funskjon som gir god tilpasning til data for væskevolumet, Vv:
def func_Vv(x, a,b, c):
    return a*x**2+b*x+c                

Vv_opt, Vv_cov = scipy.optimize.curve_fit(func_Vv, temp_vann, np.log(Vv_eksp))  # Bruker log for å tydeligere se hvor god tilpasningen er

# Plot Vv
plt.plot(temp_vann, Vv_eksp)
plt.plot(temp_vann, np.exp(func_Vv(temp_vann, *Vv_opt)), "r--")  
plt.title("Vv")
plt.xlabel("Temperartur (K)")
plt.ylabel("Volum (m^3/mol)")
plt.legend(["Eksperimentell måleserie", "Vv(t) = exp{at^2+bt+c}"])
plt.text(265, 5e-5, f"a={Vv_opt[0].round(7)}, b={Vv_opt[1].round(4)}, c={Vv_opt[2].round(7)}")
plt.show()


#------------------------------ Vg ---------------------------------------------------

# Finner funskjon som gir god tilpasning til data for gassvolumet, Vg:
def func_Vg(x, a, b,c,d,e):
    return a*x**4+b*x**3+c*x**2+d*x+e                   

Vg_opt, Vg_cov = scipy.optimize.curve_fit(func_Vg, temp_gass, np.log(Vg_eksp))  # Bruker log for å tydeligere se hvor god tilpasningen er


# Plot Vg
plt.plot(temp_gass, (Vg_eksp))
plt.plot(temp_gass, np.exp(func_Vg(temp_gass, *Vg_opt)), "r--")
plt.title("Vg")
plt.xlabel("Temperartur (K)")
plt.ylabel("Volum (m^3/mol)")
plt.legend(["Eksperimentell måleserie", "Vg(t) = exp{at^4+bt^3+ct^2+dt+e}"])
plt.text(405, 1, f"a={Vg_opt[0].round(10)}, b={Vg_opt[1].round(7)}, c={Vg_opt[2].round(4)}, d = {Vg_opt[3].round(2)}, e = {Vg_opt[4].round(2)}")
plt.show()
```

For å finne passende funskjoner brukes scipy.optimize.curve_fit. Ved å plotte den tilnærmede funskjonen samt de eksperimentelle verdiene kan man prøve seg fram til man finner en god funksjon. Ved hjelp ifra hintet til oppg 2b) fra oppgavebeskrivelsen fant vi en god tilpasning til latent varme. Først fant vi andre tilnærminger til Vv og Vg enn i koden over, men dette skapte problemer med å få en god tilpasning når vi tok i bruk Simpsons metode. For å se om tilnærmingen var nøyaktig nok, plottet vi funskjonene logaritmisk, som viste at det var en betydelig feil mellom curve_fit funksjonen og eksperimentelle verdier for Vg og Vv. Derfor brukte vi logaritmen til å finne en god tilpasning til Vv og Vg, som til slutt ga oss et mer nøyaktig resultat.
