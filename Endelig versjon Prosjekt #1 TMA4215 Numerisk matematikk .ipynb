**a)** Wish to prove that 

$ \langle e^{2\pi i k\cdot}, e^{2\pi i h\cdot} \rangle = \Biggl\{
   {\begin{array}{lr}
        1, & \text{if } h = k\\
        0, & \text{if } h\neq k
    \end{array}}
    \text{  for any } k, h \in Z, \text{ and } x\in T 
$ 

i.e. that the two functions are orthonormal. For this to be true, the two functions must be orthogonal, and of length one. 



$$ \text{Definition of an inner product of two functions: } \langle f, g\rangle = \int_{0}^{1} f(x)\overline{g(x)}\,dx \text{ where } \overline{g(x)} \text{ denotes the compelx conjugate. } $$ 

$$ \text{Thereby have that } \\ \langle e^{2\pi i k\cdot}, e^{2\pi i h\cdot} \rangle = \int_{0}^{1} e^{2\pi i kx}\overline{e^{2\pi i hx}}\,dx = \int_{0}^{1} e^{2\pi i kx}e^{-2\pi i hx}\,dx = \int_{0}^{1} e^{2\pi i x (k-h)}\,dx $$ 

$$ \text{Must consider the two cases, } h = k \text{ and } h \neq k $$ 

$$ \text{For } h = k: \int_{0}^{1} e^{2\pi i x (k-k)}\,dx = \int_{0}^{1} e^{2\pi i x (0)}\,dx = \int_{0}^{1} e^{0}\,dx = \int_{0}^{1} 1\,dx = x \bigg|_{0}^{1} = 1 - 0 =  1 $$ 


$$ \text{For } h \neq k: \int_{0}^{1} e^{2\pi i x (k-h)}\,dx = \frac{1}{2\pi i (k-h)}e^{2\pi i x (k-h)}\bigg|_{0}^{1} = \frac{1}{2\pi i (k-h)}e^{2\pi i(k-h)} - \frac{1}{2\pi i (k-h)}e^{2\pi i \cdot 0 (k-h)} = \frac{1}{2\pi i (k-h)}(e^{2\pi i})^{(k-h)} - \frac{1}{2\pi i (k-h)} \\ \text{ Using the relation that } e^{2\pi i} = 1 \implies = \frac{1}{2\pi i (k-h)}(1)^{(k-h)} - \frac{1}{2\pi i (k-h)} = \frac{1}{2\pi i (k-h)} - \frac{1}{2\pi i (k-h)} = 0 $$ 

$$ \text{Have proved that the two functions are orthonormal. }$$

$$ \implies \frac{1}{N} \Biggr( \sum_{j = 1}^{N-1} f(\frac{j}{N})e^{-2\pi i j k/N } + \frac{f(\frac{N}{N})e^{-2\pi i k N/N} + f(\frac{0}{N})e^{-2\pi i k 0 / N}}{2} \Biggr) = \frac{1}{N} \Biggr(\sum_{j = 0}^{N-1} f(\frac{j}{N}) e^{-2\pi i j k/N } +  \frac{f(1)(e^{2\pi i })^{-k} - f(0)e^{-2\pi i k 0 / N}}{2}\Biggr) = \frac{1}{N} \Biggr(\sum_{j = 0}^{N-1} f(\frac{j}{N}) e^{-2\pi i j k/N } +  \frac{f(1) - f(0)}{2}\Biggr) $$

$$ \hat f_k = \frac{1}{N} \Biggr(\sum_{j = 0}^{N-1} f(\frac{j}{N}) e^{-2\pi i j k/N }\Biggr) \implies \hat f_{k + N} =  \frac{1}{N} \Biggr(\sum_{j = 0}^{N-1} f(\frac{j}{N}) e^{-2\pi i j (k + N )/N }\Biggr) = \frac{1}{N} \Biggr(\sum_{j = 0}^{N-1} f(\frac{j}{N}) e^{-2\pi i jk/N }e^{-2\pi i jN/N }\Biggr) = \frac{1}{N} \Biggr(\sum_{j = 0}^{N-1} f(\frac{j}{N}) e^{-2\pi i jk/N }e^{-2\pi i j}\Biggr) = \frac{1}{N} \Biggr(\sum_{j = 0}^{N-1} f(\frac{j}{N}) e^{-2\pi i j k/N }\Biggr)$$


---


The sum above only gives an approximation of the Fourier coefficients, and the periodicity implies that there are only a given set of k coefficients in the periodicity interval. This implies that the Fourier coefficients for the function just are multiples of the k unique coefficients in this interval, i.e. the coefficient values are periodic. If you want use the approximation you thereby only need to calculate the k first coefficients, and reuse these. The precision of the approximation of the coefficents thereby only depend on the initial approximation, which is dependent on the number of elements one sums over. To improve the approximation one therefore would need to sum over more elements.

```python
def transform(f, N , start = 0.0): 
    
    """
       Takes N samples of the function given as agrument on the interval start + 1
       ----------------------------------------------------------------------------
       Input: 
       - f = function to sample 
       - N = number of samples 
       - start = starting point of interval, by default equal to zero 
       
       Output:
       - samp = N samples of the function on the interval start + 1 
       - x_vals = sampling points
    """
    
    # Array with sampling points 
    x_vals = np.linspace(start, start + 1, N, endpoint = False)
    
    # Sampeling function on interval 
    samp = f(x_vals)
    
    return samp, x_vals  
```

```python
# Functions defined on T = [-1/2, 1/2) as periodic functions 

# I. 
def f(x): 
    return np.sin(8*np.pi*x)

# II. 
def g(x): 
    return np.sin(32*np.pi*x) + np.cos(128*np.pi*x)

# III.
def h(x):
    return x

# IV.
def l(x): 
    return 1 - abs(x)
```

```python

def plot_n(functions, disc_points, im, title, N): 
    
    """
       Plots n plots beside each other, takes into account if function values are 
       imaginary or not 
       ----------------------------------------------------------------------------
       Input: 
       - functions = array containing n arrays with function values for functions 
                     evaluated in N discrete points 
       - disc_points = array containing n arrays with the discrete points which the 
                       functions are evaluated in 
       - im = boolean value, equals True if function values are imaginary 
       - title = string, title of plot 
       - N = array with number of samples 
    """
    
    # Initilaizing plot, adjusting placements of subplots 
    fig = plt.figure(figsize = (15, 5))
    fig.subplots_adjust(hspace = 0.4, wspace = 0.4)
    
    # Number of subplots 
    n = len(functions)
    
    # Plotting n subplots 
    for i in range(1, n + 1): 
        ax = fig.add_subplot(1, n, i)
        
        # If function values are complex 
        if im == True: 
            ax.plot(disc_points[i-1], functions[i-1].real, label = "Real") 
            ax.plot(disc_points[i-1], functions[i-1].imag, label = "Imaginary")
            ax.set_xlabel(f"N = {N[i-1]}")
            ax.grid(ls = "-.")
            ax.legend()
            
        else: 
            ax.plot(disc_points[i-1], functions[i-1])
            ax.set_xlabel(f"N = {N[i-1]}")
            ax.grid(ls = "-.")
            
    # Set title for whole plot 
    fig.suptitle(f"{title}")
    plt.show()
    
```

```python
 
def dirichlet(x, n): 
    
    """
       Calculates the Dirichlet kernel for discrete points x summed over n elements 
       -----------------------------------------------------------------------------
       Input: 
       - x = array with discrete points 
       - n = number of elements to sum over 

       Output: 
       - d + 1 = Array containing the Dirichlet kernel for each discrete point x   
    """
    
    # Variable to store Dirichlet kernel 
    d = 0 
    
    # Calculating sum with n elements  
    for i in range(0, n): 
        d += 2*np.cos(2*np.pi*i*x)
        
        
    return d + 1  # Dirichlet kernel 
        
```

**f)** An filter that does the oposite of the Dirchlet kernel, is an filter that retains the high frequencies of an signal, i.e. an high-pass filter. This can be obtained by taking the inverse of the Dirichlet kernel. The Dirichlet kernel $D_n$ can be represented as an vector with lenght N > n with n coefficintes c_k = 1 on either side of the middle of vector, i.e. the zero point. The rest of the elements are simply just zeros. One easy way to get a kernel that does the oposite of the Dirichlet kernel is then to make an array of length N containing only ones, and then subtracting the array representation of the Dirichlet kernel. The resulting array is then a kernel which retains high frequency parts of a signal.

```python

def multivariate_fourier_func(N1, N2, k, start):
    
    """
       Applys the Fourier transform to two dimensional function 
       Plots function gray-scale image of function in 2D as well as the absolute 
       value of the DFT as a gray-scale image
       --------------------------------------------------------------------------
       Input:
       - N1, N2 = number of discrete points on the x- and y-axis respectively 
       - k = vector in R^2 
       - start = start of interval   
    """
    
    
    x_disc = np.linspace(start, start + 1, N1)    # Discrete points on x- and y-axis 
    y_disc = np.linspace(start, start + 1, N2)
    
    x, y = np.meshgrid(x_disc, -y_disc)           # Two dimensional coordinate system
    
    
    grating = 1  + 1/2 * np.sin(2*np.pi * (k[0]*x + k[1]*y))   # Function evaluated in every discerte point in
                                                               # the defined coordinate system 
    
    
    disc_fourier = scipy.fft.fft2(grating)          # FFT
    disc_shift =  scipy.fft.fftshift(disc_fourier)  # Shifting FFT 
    
    
    # Maximum value in array for normalizing 
    max_val = abs(np.amax(disc_shift))
    
    
    # Initializing figure 
    plt.figure(figsize = (15,5))
    
    # Plotting gray-scale image of function in 2D 
    plt.set_cmap("gray")
    ax1 = plt.subplot(121)
    ax1.set_title(f"Image of function in 2D\n k ={k}")
    plt.imshow(grating)
    
    # Plotting the absolute value of DFT as gray-scale image 
    ax2 = plt.subplot(122)
    ax2.set_title("Absolute value of DFT")
    plt.imshow(abs(disc_shift/max_val))     # Normalized 
    plt.show()
    
```

```python
def padding(kernel, x, y): 
    
    """
       Padds matrix with zeros to obtain dimension x*y 
       -------------------------------------------------
       Input:
       - kernel = matrix of arbitary size 
       - x, y = x,y dimension of image 
       Output 
       - grating = kernel padded with zeros to obtain same size as image 
    """
    
    # Creating filter the same size as the image 
    grating = np.zeros((x, y))                           # Zero padding to ensure filter is the same size as image 
    grating[:kernel.shape[0], :kernel.shape[1]] = kernel 
    
    return grating
```

```python
def convolution_2d(kernel, img, v2 = False): 
    
    """
       Padds matrix with zeros to obtain dimension x*y 
       -------------------------------------------------
       Input:
       - kernel = matrix containing kernel, same dimension as the image 
       - img = image  
       Output 
       - conv = convolution of image and kernel 
    """
    
    # Transforming image 
    fft_img = scipy.fft.fft2(img)
    
    if v2 == True: 
        shift_img = np.fft.fftshift(fft_img)
        conv = scipy.fft.ifft2((np.fft.fftshift(kernel*shift_img)))
    else: 
    
        # Transforming filter 
        fft_filter = scipy.fft.fft2(kernel)
    
        # Finding the convolution 
        conv = scipy.fft.ifft2(fft_img*fft_filter)
    
    return conv 
    
```

```python
def plot_conv(conv, title): 
    
    """
       Plots the 2D convolution of image
       ----------------------------------
       Input:
       - conv = array containing arrays with convolutions  
       - title = array containing names of filters image have been convolved with 
    """
    
    # Initilaizing plot, adjusting placements of subplots 
    fig = plt.figure(figsize = (15, 5))
    fig.subplots_adjust(hspace = 0.4, wspace = 0.4)
    
    n = len(conv)
    
    # Plotting convolutions beside each other 
    for i in range(1, n+1): 
        ax = fig.add_subplot(1, n, i)
        ax.imshow(conv[i-1].real)
        ax.set_xticks([])
        ax.set_yticks([])
        ax.set_xlabel(f"Convolution with filter {title[i-1]}")
        
        
```

```python
def dirichlet_2d(n, x_dim, y_dim, start): 
    
    """  
       Calculates the 2D Dirichlet kernel 
       -----------------------------------
       Input:
       - n = number of elements to sum over
       - x_dim, y_dim = x- and y- dimension of image 
       - start = start of interval 
       
       Output: 
       - 2D-dirichlet kernel 
       
    """
    
    x_disc = np.linspace(start, start + 1, x_dim)
    y_disc = np.linspace(start, start + 1, y_dim)
    
    x, y = np.meshgrid(x_disc, y_disc)
    
    dirichlet_x = dirichlet(x, n)
    dirichlet_y = dirichlet(y, n)
    
    return dirichlet_x * dirichlet_y
```

```python

def dirichlet_new(n, x_dim, y_dim):
    
    """  
       Calculates the 2D Dirichlet kernel, consturcted using only zeros and ones 
       i.e. defined in the frequency domain. 
       --------------------------------------------------------------------------
       Input:
       - n = number of elements to sum over
       - x_dim, y_dim = x- and y- dimension of image  
       
       Output: 
       - 2D-dirichlet kernel 
       
    """
    
    x = np.zeros(x_dim)
    y = np.zeros(y_dim)
    
    x[int(x_dim/2)-n-1: int(x_dim/2)+n] = 1
    y[int(y_dim/2)-n-1: int(y_dim/2)+n] = 1
    
    d = np.zeros((x_dim, y_dim))
    
    for i in range(x_dim): 
        for j in range(y_dim): 
            d[i][j] = x[i]*y[j]
            
    return d 
```

```python
def image_transform(image_file, n, start, name): 
    
    """  
       Loads and displays image given as argument. Applys FFT to image, and plots amplitude and the 
       convolution of the image with the Dirichlet kernel beside the image 
       -------------------------------------------------------------------------------------------------
       Input:
       - image_file = name of file 
       - n = number of elements to sum over 
       - start = start of interval 
       - name = name of image 
    """
    
    # Opening image from file 
    img = Image.open(image_file)
    
    # Size of image 
    x_img , y_img = img.size
    
    # FFT of image 
    fft_img = scipy.fft.fft2(img)
    
    # Shifted FFT of image 
    fftshift_img = scipy.fft.fftshift(fft_img)
    
    # Dirichlet kernel with approprioate dimension for given image 
    dirichlet_kernel = dirichlet_2d(n, x_img, y_img, start)
    
    # Dirichlet kernel, second implementation 
    dirichlet_kernel_v2 = dirichlet_new(n, x_img, y_img).T
    
    # Applying convolution with the Dirichlet kernel 
    conv_dirichlet = convolution_2d(dirichlet_kernel, img)
    
    conv_dirichlet_v2 = convolution_2d(dirichlet_kernel_v2, img, True)
    

    # Initializing figure 
    plt.figure(figsize = (15,5))
    
    # Plotting gray-scale image of function in 2D 
    plt.set_cmap("gray")
    ax1 = plt.subplot(141)
    ax1.set_xticks([])
    ax1.set_yticks([])
    plt.imshow(img)
    ax1.set_title(f"{name} image")
    
    # Plotting the absolute value of DFT as gray-scale image 
    ax2 = plt.subplot(142)
    ax2.set_title("Absolute value of DFT of image")
    ax2.set_xticks([])
    ax2.set_yticks([])
    plt.imshow(abs(fftshift_img), norm = LogNorm())     
   
    
    
    if name == "Yarimton": 
             
            # Must invert back image
            inverted = 1 -(scipy.fft.fftshift(conv_dirichlet).real)
            inverted_v2 = 1 - (abs(conv_dirichlet_v2))
            
            # Plotting convolution of image with the Dirichlet kernel 
            ax3 = plt.subplot(143)
            ax3.set_title(f"Convolution with Dirichlet kernel\nn={n}")
            ax3.set_xticks([])
            ax3.set_yticks([])
            plt.imshow(inverted) 
            
            # Plotting convolution of image with the Dirichlet kernel 
            ax4 = plt.subplot(144)
            ax4.set_title(f"Convolution with Dirichlet kernel\nn={n}, Second implementation")
            ax4.set_xticks([])
            ax4.set_yticks([])
            plt.imshow(inverted_v2)
    else:
        
        # Plotting convolution of image with the Dirichlet kernel 
        ax3 = plt.subplot(143)
        ax3.set_title(f"Convolution with Dirichlet kernel\nn={n}")
        ax3.set_xticks([])
        ax3.set_yticks([])
        plt.imshow(scipy.fft.fftshift(conv_dirichlet).real) 
    
        # Plotting convolution of image with the Dirichlet kernel 
        ax4 = plt.subplot(144)
        ax4.set_title(f"Convolution with Dirichlet kernel\nn={n}, Second implementation")
        ax4.set_xticks([])
        ax4.set_yticks([])
        plt.imshow(abs(conv_dirichlet_v2))

               
```
